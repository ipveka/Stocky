### Backtesting

```
# Get data


data = get_crypto_data("BTC/USDT", "2018-12-01", "2019-12-31")
data.head()
```

```
# SMAC

smac = backtest('smac', data, fast_period = 15, slow_period = 40, verbose = False)
```

```
# Optim SMAC

smac_optim = backtest("smac", data, fast_period = range(15, 30, 3), slow_period = range(40, 55, 3), verbose = False)

print(smac_optim[['fast_period', 'slow_period', 'final_value']].head())
```

```
# RSI

backtest('rsi', data, rsi_period = 14, rsi_upper = 70, rsi_lower = 30)
```

```
# SMAC Crossover

backtest('smac', data, fast_period = 10, slow_period=  30)
```

```
# EMAC

backtest('emac', data, fast_period = 10, slow_period = 30)
```

```
# MACD

backtest('macd', data, fast_period = 12, slow_period = 26, signal_period = 9, sma_period = 30, dir_period = 10)
```

```
# Bollinger bands

backtest('bbands', data, period = 20, devfactor = 2.0)
```

```
# Multi

# Use single set of parameters

strats = { 
    "smac": {"fast_period": 35, "slow_period": 50}, 
    "rsi": {"rsi_lower": 30, "rsi_upper": 70} 
} 
res = backtest("multi", data, strats = strats)
res.shape

# Use auto grid search

strats_opt = { 
    "smac": {"fast_period": 35, "slow_period": [40, 50]}, 
    "rsi": {"rsi_lower": [15, 30], "rsi_upper": 70} 
} 

res_opt = backtest("multi", data, strats = strats_opt)
res_opt.shape
```
